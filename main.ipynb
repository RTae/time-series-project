## EEGImageNetDataset usage examples

This shows the recommended explicit constructor and the compatibility `from_args` style.

```python
from argparse import Namespace
from torch.utils.data import DataLoader
from src.dataset import EEGImageNetDataset

# 1) Recommended: explicit constructor
train_data = EEGImageNetDataset(
    dataset_dir="data",
    subject=-1,         # -1 means all subjects
    granularity="all", # or "coarse", "fine", "fine0", "fine1", ...
    eeg_window=(40, 440),
)

print("dataset size:", len(train_data))
x, y = train_data[0]
print("one sample eeg shape:", tuple(x.shape), "label:", y)

loader = DataLoader(train_data, batch_size=8, shuffle=True)
batch_x, batch_y = next(iter(loader))
print("batch eeg shape:", tuple(batch_x.shape), "batch label shape:", tuple(batch_y.shape))
```

```python
# 2) Compatibility: from_args (if you still use argparse.Namespace)
args = Namespace(dataset_dir="data", subject=-1, granularity="all")
compat_data = EEGImageNetDataset.from_args(args)
print("compat dataset size:", len(compat_data))
```
